- **COMPARACI√ìN**
    - **Gr√°fico de Barras:** Utilizado para comparar puntos de datos individuales a trav√©s de categor√≠as. Cada barra representa una categor√≠a y la altura o longitud de la barra es proporcional al valor o conteo de esa categor√≠a.
    - **Gr√°fico de Barras Agrupadas:** Una variaci√≥n del gr√°fico de barras que muestra dos o m√°s grupos lado a lado para la comparaci√≥n a trav√©s de categor√≠as y subcategor√≠as.
    - **Gr√°fico de L√≠neas:** Ideal para mostrar tendencias a lo largo del tiempo. Cada punto en la l√≠nea representa un valor de datos en un per√≠odo de tiempo espec√≠fico.
    - **Diagrama de Dispersi√≥n:** Perfecto para mostrar la relaci√≥n entre dos variables num√©ricas, permitiendo la identificaci√≥n de correlaciones, patrones o valores at√≠picos.

- **DISTRIBUCI√ìN**
    - **Histograma:** Proporciona una visualizaci√≥n de la distribuci√≥n de frecuencias subyacente de un conjunto de datos num√©ricos continuos o discretos.
    - **Swarmplot:** Representa todos los puntos de datos a lo largo de un eje, √∫til para mostrar la distribuci√≥n de un conjunto de datos e identificar posibles valores at√≠picos.
    - **Diagrama de Caja:** Resume la distribuci√≥n de un conjunto de datos basado en el resumen de cinco n√∫meros: m√≠nimo, primer cuartil (Q1), mediana, tercer cuartil (Q3) y m√°ximo.
    - **Diagrama de Viol√≠n:** Combina las propiedades de los diagramas de caja y los gr√°ficos de densidad kernel, proporcionando un resumen de la distribuci√≥n de los datos.

- **PARTE DE UN TODO**
    - **Gr√°fico Circular:** Representa proporciones o porcentajes entre categor√≠as, dividiendo un c√≠rculo en segmentos proporcionales.
    - **Gr√°fico de Columnas Apiladas:** Muestra la comparaci√≥n de los totales individuales de subcategor√≠as con el total general.

- **RELACIONES**
    - **Diagrama de Dispersi√≥n:** Ideal para representar la relaci√≥n entre dos variables num√©ricas, permitiendo la detecci√≥n de correlaciones, patrones o valores at√≠picos.
    - **Gr√°fico de L√≠neas:** √ötil para mostrar tendencias a lo largo del tiempo, mostrando la relaci√≥n entre dos variables num√©ricas a lo largo de un lapso de tiempo continuo.
    - **Mapa de Calor:** Proporciona una representaci√≥n gr√°fica coloreada de datos donde los valores individuales se representan como colores, √∫til para representar correlaciones o magnitudes.
    - **Pairplot:** Muestra relaciones por pares en un conjunto de datos, ideal para visualizar r√°pidamente distribuciones y correlaciones.

- **TENDENCIAS**
    - **Gr√°fico de L√≠neas:** Efectivo para representar tendencias a lo largo de un intervalo de tiempo continuo o discreto.
    - **Gr√°fico de √Årea:** Similar a los gr√°ficos de l√≠neas pero con el √°rea bajo la l√≠nea rellenada, √∫til para representar el efecto acumulativo o total a lo largo del tiempo.
    - **Gr√°fico de Columnas:** √ötil para comparar categor√≠as discretas de datos, mostrando las diferencias entre ellas a trav√©s de barras verticales.

#### seaborn: barplot

El barplot de Seaborn es una visualizaci√≥n vers√°til que nos permite representar la tendencia central de una variable num√©rica usando barras rectangulares. A diferencia del countplot, que se utiliza para datos categ√≥ricos, el barplot est√° dise√±ado para datos num√©ricos. Proporciona una representaci√≥n visual del valor medio estimado para cada categor√≠a dentro de una variable categ√≥rica, junto con una medida de incertidumbre representada por barras de error.

**Descripci√≥n:**
- El barplot es particularmente √∫til cuando quieres comparar los valores medios de una variable num√©rica a trav√©s de diferentes categor√≠as.
- Cada barra rectangular en el gr√°fico representa una categor√≠a, y la altura de la barra corresponde a la tendencia central estimada (t√≠picamente la media) de la variable num√©rica para esa categor√≠a.
- Se incluyen barras de error para indicar la incertidumbre o variabilidad alrededor del valor estimado. Estas barras de error pueden personalizarse para representar varias medidas estad√≠sticas.

`average passengers by year‚ùì`

```python
# Display the first few rows of the 'flights' DataFrame
flights.head()
```

```python
# Grouping the data by 'year' and calculating the mean of 'passengers' rounded to 3 decimal places
flights.groupby("year")["passengers"].mean().round(3)
```

```python
# Creating a barplot to visualize the average number of passengers by year
# The 'ci=None' parameter is used to exclude confidence intervals
barplot = sns.barplot(x="year", y="passengers", data=flights, ci=None)
```

`total passengers by year‚ùì`

```python
# Creating a barplot to visualize the total number of passengers by year
# 'x' represents the 'year' data, and 'y' represents the 'passengers' data
# 'palette="magma"' sets the color palette to 'magma'
# 'estimator=sum' calculates the sum of passengers for each year
# 'ci=None' is used to exclude confidence intervals
sns.barplot(x=flights["year"], y=flights["passengers"], palette="magma", estimator=sum, ci=None);
```

```python
# Count the number of penguins in each species and display the counts
species_counts = penguins["species"].value_counts()

# Create a bar plot with specific options:
# - 'kind="bar"' specifies the type of plot as a bar chart
# - 'color="salmon"' sets the color of the bars to salmon
# - 'title="Number of penguins by species"' adds a title to the plot
species_counts.plot(kind="bar", color="salmon", title="Number of penguins by species")

# Rotate the x-axis labels slightly for better readability
plt.xticks(rotation=45)
```

### Column: grouped bar chart

En la visualizaci√≥n de datos, un gr√°fico de barras agrupadas nos permite presentar datos con un nivel adicional de detalle y complejidad. Logramos esto utilizando el par√°metro `hue`, una herramienta poderosa proporcionada por seaborn. Cuando usamos el par√°metro `hue` en un gr√°fico de barras agrupadas, efectivamente multiplica el n√∫mero de barras por los valores distintos de una variable espec√≠fica que especificamos.

En este ejemplo, estamos trabajando con el conjunto de datos del Titanic de seaborn. Al incorporar el par√°metro `hue`, podemos segmentar y diferenciar las barras dentro de cada categor√≠a, proporcionando informaci√≥n sobre m√∫ltiples dimensiones de nuestros datos. Este enfoque es particularmente √∫til cuando queremos visualizar relaciones y comparaciones entre diferentes subgrupos o variables categ√≥ricas, haciendo que nuestros gr√°ficos sean m√°s informativos e intuitivos.

En la visualizaci√≥n de datos, personalizar los colores de tus gr√°ficos y tramas es una manera poderosa de mejorar su atractivo visual y transmitir informaci√≥n de manera efectiva. Matplotlib, una biblioteca de Python ampliamente utilizada para crear visualizaciones est√°ticas, animadas e interactivas, ofrece varias opciones para personalizar los colores en tus gr√°ficos.

Para personalizar los colores de las barras en un gr√°fico de barras usando Matplotlib, puedes usar el par√°metro `color`. Este par√°metro te permite especificar los colores de las barras individuales en el gr√°fico, proporcionando flexibilidad para lograr el esquema de color deseado.

Aqu√≠ te mostramos c√≥mo puedes personalizar los colores de un gr√°fico de barras en Matplotlib:

1. Importa las bibliotecas requeridas, t√≠picamente `matplotlib.pyplot` para la creaci√≥n de gr√°ficos.
2. Crea tu conjunto de datos, definiendo las categor√≠as y los valores correspondientes.
3. Crea un gr√°fico de barras b√°sico usando `plt.bar()`.
4. Agrega el par√°metro `color` a tu funci√≥n `plt.bar()` y especifica los colores que deseas usar. Puedes usar colores nombrados o proporcionar valores RGB.

Puedes experimentar con diferentes colores y combinaciones para crear gr√°ficos de barras visualmente atractivos e informativos.
```python
# Group the penguins dataframe by two variables: "species" and "sex," and then count the occurrences of each combination.
# This gives us a count of penguins for each species and gender.
penguin_counts = penguins.groupby(["species", "sex"])["sex"].count()

# Create a bar chart to visualize the penguin counts.
# Use the "kind" parameter to specify that we want a bar chart.
# Additionally, set the colors for the bars using the "color" parameter.
# The first color "slategray" corresponds to one gender, and the second color "coral" corresponds to the other gender.
penguin_counts.plot(kind="bar", color=["slategray", "coral"])

# Add axis labels, titles, and legends if necessary for clarity.
plt.xlabel("Species and Gender", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Penguin Counts by Species and Gender", fontsize=20)

# Show the plot.
plt.show()
```

**Stacking y unstacking** son operaciones en Pandas que te permiten remodelar tus datos entre formatos largos (stacked) y anchos (unstacked):

- **Stacking** t√≠picamente implica mover datos de columnas a filas, creando un √≠ndice jer√°rquico (de m√∫ltiples niveles). Es √∫til para hacer que tus datos sean largos y estrechos.

- **Unstacking** hace lo contrario, moviendo datos de filas de vuelta a columnas, esencialmente pivotando tus datos. Esto es √∫til para crear datos en formato ancho, que a menudo se utilizan en visualizaciones como gr√°ficos de barras agrupadas.

Entonces, para agrupar barras en un gr√°fico de barras, desapilaremos nuestro DataFrame, lo que reorganizar√° los datos en un formato que es adecuado para crear barras agrupadas. Puedes aprender m√°s sobre el m√©todo `unstack()` [aqu√≠](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.unstack.html).

Para una comprensi√≥n m√°s profunda del multi-√≠ndice en Pandas, puedes referirte a la documentaci√≥n oficial sobre [indexaci√≥n avanzada](https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html).

### Gr√°fico de L√≠neas

Un gr√°fico de l√≠neas es una herramienta de visualizaci√≥n popular utilizada para mostrar puntos de datos a lo largo de un intervalo continuo o per√≠odo de tiempo. Es particularmente √∫til para mostrar tendencias, patrones y variaciones en los datos. Los gr√°ficos de l√≠neas se usan com√∫nmente en an√°lisis y visualizaci√≥n de datos para representar datos con un aspecto temporal, como precios de acciones, cambios de temperatura o cifras de ventas a lo largo del tiempo.

**Caracter√≠sticas Clave de los Gr√°ficos de L√≠neas:**

- **Eje X**: T√≠picamente representa la variable independiente, como el tiempo, fechas o valores num√©ricos que tienen un orden continuo.

- **Eje Y**: Representa la variable dependiente, que puede ser cualquier valor num√©rico o cantidad que desees visualizar. Puede ser continuo o discreto.

- **Puntos de Datos**: Los puntos de datos se trazan como puntos o marcadores en valores espec√≠ficos a lo largo de los ejes X e Y. Estos puntos est√°n conectados por l√≠neas, que crean la caracter√≠stica "l√≠nea" en el gr√°fico.

- **L√≠neas**: Las l√≠neas que conectan puntos de datos ayudan a visualizar las tendencias y patrones en los datos. Facilitan la identificaci√≥n de si los valores est√°n aumentando, disminuyendo o manteni√©ndose relativamente constantes a lo largo del tiempo o en diferentes categor√≠as.

- **Casos de Uso**: Los gr√°ficos de l√≠neas se utilizan com√∫nmente para datos de series de tiempo, an√°lisis de precios de acciones, pron√≥sticos de ventas y cualquier escenario donde desees visualizar tendencias de datos.

- **Personalizaci√≥n**: Los gr√°ficos de l√≠neas pueden personalizarse con varias opciones, como colores, marcadores, etiquetas y anotaciones para mejorar la presentaci√≥n y claridad del gr√°fico.

- **Interactividad**: En bibliotecas de visualizaci√≥n de datos interactivas como Plotly o Bokeh, los gr√°ficos de l√≠neas pueden hacerse interactivos, permitiendo a los usuarios hacer zoom, desplazarse o pasar el cursor sobre puntos de datos para obtener m√°s informaci√≥n.

Los gr√°ficos de l√≠neas son una herramienta poderosa para la exploraci√≥n y presentaci√≥n de datos. Ayudan a los analistas y tomadores de decisiones a entender c√≥mo cambian los datos a lo largo del tiempo o en diferentes categor√≠as, haci√©ndolos una parte esencial del kit de herramientas de visualizaci√≥n de datos.

En la visualizaci√≥n de datos, entender la distribuci√≥n de tus datos es esencial. Te ayuda a obtener informaci√≥n sobre c√≥mo se dispersan tus datos y si siguen un patr√≥n particular. Las siguientes t√©cnicas se utilizan com√∫nmente para explorar la distribuci√≥n de datos:

- **Histogramas:** Los histogramas proporcionan una representaci√≥n visual de la frecuencia o conteo de puntos de datos dentro de intervalos o bins espec√≠ficos. Son √∫tiles para comprender la forma y tendencias centrales de datos num√©ricos.

- **Boxplots:** Los boxplots, tambi√©n conocidos como diagramas de caja y bigotes, muestran la distribuci√≥n de datos a trav√©s de cuartiles. Ayudan a identificar valores at√≠picos y proporcionan un resumen conciso de la dispersi√≥n de los datos.

- **Estimaci√≥n de Densidad Kernel (KDE):** KDE es una t√©cnica para estimar la funci√≥n de densidad de probabilidad de una variable aleatoria continua. Ayuda a visualizar la distribuci√≥n suave de puntos de datos y es particularmente √∫til para comprender la distribuci√≥n subyacente de los datos.

Cada uno de estos m√©todos tiene sus fortalezas y es adecuado para diferentes tipos de exploraci√≥n de datos. Elige el que mejor se ajuste a tus necesidades anal√≠ticas al examinar la distribuci√≥n de tus datos.

### Histogramas

Los histogramas son representaciones gr√°ficas que se asemejan a los gr√°ficos de barras verticales. Sin embargo, sirven a un prop√≥sito distinto en la visualizaci√≥n de datos. Los histogramas son particularmente √∫tiles para comprender la distribuci√≥n de frecuencia subyacente de un conjunto de datos, ya sea que consista en valores discretos o continuos medidos en una escala de intervalo.

Caracter√≠sticas clave y beneficios de los histogramas incluyen:

- **Distribuci√≥n de Frecuencia:** Los histogramas proporcionan informaci√≥n sobre la distribuci√≥n de puntos de datos dentro de intervalos o bins especificados. Al visualizar c√≥mo se distribuyen los datos a trav√©s de estos bins, puedes observar patrones y tendencias.

- **Asimetr√≠a y Curtosis:** Los histogramas te permiten evaluar propiedades como la asimetr√≠a (si los datos est√°n sesgados hacia la izquierda o derecha) y la curtosis (que mide la concentraci√≥n de valores alrededor de la regi√≥n central de la distribuci√≥n). ### Boxplot

Un boxplot, tambi√©n conocido como diagrama de caja y bigotes, es una herramienta valiosa para visualizar la distribuci√≥n de un conjunto de datos e identificar medidas estad√≠sticas clave. Proporciona un resumen conciso de la tendencia central de los datos, la dispersi√≥n y la presencia de valores at√≠picos.

Componentes clave de un boxplot:
- **Mediana (Q2)**: La l√≠nea dentro de la caja representa la mediana, que es el valor medio del conjunto de datos cuando se ordena en orden ascendente. Divide los datos en dos mitades iguales, con el 50% de los valores por debajo y el 50% por encima de ella.

- **Rango Intercuartil (IQR)**: La caja encierra el IQR, que es el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3). El IQR abarca el 50% central de los datos y proporciona una medida de la dispersi√≥n de los datos.

- **Bigotes**: Los bigotes se extienden desde los bordes de la caja hasta los valores m√≠nimos y m√°ximos dentro de un rango definido. Los valores at√≠picos, si est√°n presentes, se muestran como puntos individuales m√°s all√° de los bigotes.

- **Valores At√≠picos**: Los valores at√≠picos son puntos de datos que caen significativamente fuera de los bigotes. T√≠picamente se muestran como puntos individuales y pueden indicar valores inusuales o extremos en el conjunto de datos.

Los boxplots son valiosos para comparar distribuciones, identificar asimetr√≠a y detectar posibles valores at√≠picos en los datos. Proporcionan un resumen visual de las propiedades estad√≠sticas de los datos, facilitando la comprensi√≥n de caracter√≠sticas esenciales de un conjunto de datos.

Para crear un boxplot en Python, puedes usar bibliotecas como Matplotlib o Seaborn, especificando la variable que deseas visualizar. Los boxplots son una herramienta poderosa para el an√°lisis exploratorio de datos y para obtener insights sobre la distribuci√≥n de tus datos.

### ViolinPlot

Un Violin Plot es una t√©cnica de visualizaci√≥n de datos vers√°til que combina elementos de un box plot con un gr√°fico de densidad de kernel rotado. Es particularmente √∫til para visualizar la distribuci√≥n y densidad de probabilidad de un conjunto de datos a trav√©s de diferentes categor√≠as o variables.

Caracter√≠sticas clave de un Violin Plot:
- **Forma Sim√©trica**: El n√∫cleo de un Violin Plot se asemeja a una forma sim√©trica de viol√≠n rotado. Esta forma representa la densidad de probabilidad estimada de los datos en diferentes valores a lo largo del eje y.

- **Box Plot Interior**: Similar a un box plot, una caja central se dibuja dentro de la forma de viol√≠n. Esta caja denota el rango intercuartil (IQR) y la mediana de los datos. El ancho de la caja es proporcional a la densidad de puntos de datos en ese valor.

- **Extended Tails**: Las colas del violin plot se extienden hacia afuera, indicando el rango de valores de datos. El ancho de las colas corresponde a la densidad de puntos de datos en diferentes valores.

Los Violin Plots son beneficiosos para comparar la distribuci√≥n de datos a trav√©s de m√∫ltiples categor√≠as o grupos, especialmente cuando se trata de conjuntos de datos complejos. Proporcionan informaci√≥n sobre la tendencia central de los datos, la dispersi√≥n y la presencia de m√∫ltiples modos o picos dentro de la distribuci√≥n.

Al visualizar tanto el box plot como la estimaci√≥n de densidad de kernel, los Violin Plots ofrecen una vista comprensiva de las propiedades estad√≠sticas de los datos. Esto los convierte en una herramienta valiosa en el an√°lisis exploratorio de datos y la visualizaci√≥n de datos.

## Entendiendo las Funciones de NumPy en la Creaci√≥n de Heatmaps

En el contexto de la creaci√≥n de heatmaps, NumPy, una poderosa biblioteca de c√≥mputo num√©rico en Python, ofrece herramientas valiosas para manipular datos antes de visualizarlos. Dos funciones esenciales de NumPy com√∫nmente utilizadas en la generaci√≥n de heatmaps son `np.triu` y `np.ones_like`.

- [**Documentaci√≥n de `np.triu`**](https://numpy.org/doc/stable/reference/generated/numpy.triu.html): Esta funci√≥n, abreviatura de "triangular upper" (parte triangular superior), se utiliza para crear una m√°scara booleana que retiene solo la parte superior triangular de una matriz mientras establece la parte inferior en `False`. Es particularmente √∫til al visualizar matrices de correlaci√≥n o datos sim√©tricos similares.

- [**Documentaci√≥n de `np.ones_like`**](https://numpy.org/doc/stable/reference/generated/numpy.ones_like.html): Esta funci√≥n genera un arreglo de unos con la misma forma y tipo de datos que un arreglo de entrada especificado. A menudo se emplea para crear una matriz inicial con valores `True` o `1` antes de aplicar una m√°scara.

En el contexto de la creaci√≥n de heatmaps, estas funciones de NumPy juegan un papel crucial en mejorar la visualizaci√≥n de matrices de correlaci√≥n. Al usar `np.triu` en combinaci√≥n con `np.ones_like`, podemos enfocarnos en visualizar solo una mitad de la matriz de correlaci√≥n, t√≠picamente la porci√≥n triangular superior. Esto evita informaci√≥n redundante y resulta en un heatmap m√°s conciso e interpretable, especialmente cuando se trata con datos sim√©tricos como los coeficientes de correlaci√≥n.

Al entender c√≥mo usar estas funciones de NumPy en la creaci√≥n de heatmaps, puedes adaptar efectivamente tus visualizaciones para resaltar las relaciones esenciales dentro de tus datos, facilitando la extracci√≥n de insights y conclusiones significativas.

```python
sns.__version__
```

```python
# Creating a boolean mask for the upper triangular part of the correlation matrix
mask = np.triu(np.ones_like(corr, dtype=bool))

# Creating a color map for the heatmap using seaborn's diverging_palette
color_map = sns.diverging_palette(0, 10, as_cmap=True)

# Creating a heatmap using seaborn
sns.heatmap(corr,  
            mask=mask,               # Applying the upper triangular mask to hide redundant information
            cmap=color_map,          # Setting the color map for the heatmap
            square=True,             # Making the heatmap square
            linewidth=0.5,           # Setting the linewidth between cells
            vmax=1,                  # Defining the maximum color scale value
            vmin=-1,                 # Defining the minimum color scale value
            cbar_kws={"shrink": .5}  # Customizing the colorbar appearance
);

# Displaying the heatmap
plt.show()
```

## Resumen: Visualizaciones de An√°lisis Exploratorio de Datos (EDA)

En esta secci√≥n de visualizaci√≥n de EDA, exploramos varios tipos de visualizaciones para obtener insights de nuestros conjuntos de datos. Estas visualizaciones nos ayudan a entender las distribuciones de datos, las relaciones entre variables y m√°s. Repasemos las visualizaciones clave que cubrimos:

### Visualizaciones de Datos Categ√≥ricos:

- **Gr√°ficos de Barras:** Se utilizan para visualizar conteos de datos categ√≥ricos.
- **Gr√°ficos de Barras Agrupadas:** Muestran conteos de datos categ√≥ricos con agrupaci√≥n.
- **Gr√°ficos de Barras Apiladas:** Muestran conteos de datos con apilamiento.
- **Gr√°ficos de Conteo:** Muestran el conteo de datos categ√≥ricos usando barras.

### Visualizaciones de Datos Num√©ricos:

- **Histogramas:** Visualizan la distribuci√≥n de datos num√©ricos.
- **Estimaci√≥n de Densidad Kernel (KDE):** Grafica la funci√≥n de densidad de probabilidad sobre los histogramas.
- **Box Plots:** Muestran cuartiles e identifican valores at√≠picos.
- **Violin Plots:** Combinan box plots y KDE para visualizar la distribuci√≥n de datos.

### Visualizaciones de Relaciones:

- **Gr√°ficos de Dispersi√≥n:** Muestran relaciones entre dos variables num√©ricas.
- **Gr√°ficos de Pares:** Muestran relaciones por pares entre m√∫ltiples variables.
- **Heatmaps:** Visualizan correlaciones entre variables num√©ricas.

### Visualizaciones de Parte de un Todo:

- **Gr√°ficos Circulares:** Representan partes de un todo, pero pueden ser menos efectivos para datos cuantitativos.
- **Gr√°ficos de Barras Apiladas:** Muestran partes de un todo con barras apiladas.
- **Gr√°ficos de Columnas Apiladas:** Presentan partes de un todo con columnas apiladas.

Estas visualizaciones nos permiten explorar nuestros conjuntos de datos, identificar patrones y obtener insights para guiar el an√°lisis y la toma de decisiones posteriores.

Recuerda elegir el tipo de visualizaci√≥n m√°s apropiado basado en tus datos y los insights que deseas extraer.
